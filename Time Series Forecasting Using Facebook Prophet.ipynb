# Predicting Future Avocado Prices Using Facebook Prophet

In this project, we leverage the power of Facebook Prophet to forecast future avocado prices. Facebook Prophet is an exceptional open-source tool specifically designed for time series forecasting. With its capability to analyze daily, weekly, and yearly seasonality patterns, as well as account for holiday effects, Prophet equips us with invaluable insights into price trends.

Prophet offers fast and accurate forecasting with minimal data preprocessing. It handles missing data and outliers effectively, providing reliable predictions in seconds.

# Prepare Data to Apply to Facebook Prophet

```python
avocado_df
```

```python
avocado_prophet_df = avocado_df[['Date', 'AveragePrice']]
```

```python
avocado_prophet_df
```

```python
# rename columns 'Date' to ds and 'AveragePrice' to 'y', which is the format Prophet will read
avocado_prophet_df = avocado_prophet_df.rename(columns = {'Date': 'ds' , 'AveragePrice' : 'y'})
```

```python
avocado_prophet_df
# y is the value we are trying to forecast
```

# Develop Model and Make Predictions

```python
m = Prophet() # instantiate object
m.fit(avocado_prophet_df) # fit model with training data
```

```python
# forcast into the future
future = m.make_future_dataframe(periods = 365) # a year
forecast = m.predict(future)
```

```python
forecast
# trend of prices into the future including upper and lower bounderies (the confidence of that trend)
```

```python
figure = m.plot(forecast, xlabel = 'Date', ylabel = 'Price')
```

The black dots represent historaical data and the blue is our forecast.
We can now plan accordingly if we predict peaks or dips in demand.

```python
# performance during year and month
figure2 = m.plot_components(forecast)
```

First graph: Demand went down then up, and expectation is to keep going down.

Second graph: Prices will peak around November and dip in January.

# Develop Model and Make Predictions for Specific Region

```python
# dataframes creation for both training and testing datasets
avocado_df = pd.read_csv('avocado.csv')
```

```python
# select specific region
avocado_df_sample = avocado_df[avocado_df['region']=='West']
```

```python
# sort by time oldest to most recent
avocado_df_sample = avocado_df_sample.sort_values('Date')
```

```python
plt.plot(avocado_df_sample['Date'], avocado_df_sample['AveragePrice'])
# higher prices, higher peak at 2.5
```

```python
# condition data to feed into Prophet
avocado_df_sample = avocado_df_sample.rename(columns = {'Date' : 'ds' , 'AveragePrice' : 'y'})
```

```python
m = Prophet() # instantiate object
m.fit(avocado_df_sample)
```

```python
# forcast into the future
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
```

```python
figure = m.plot(forecast, xlabel='Date', ylabel='Price')
```

```python
figure3 = m.plot_components(forecast)
```

First graph: Demand for avocados in the West is expected to keep climbing.

Second graph: Prices are expected to peak October - November, and then dip by January.

Conclusion:

Our analysis reveals a contrasting trend in avocado demand. Despite a nationwide decline, avocado demand is expected to rise in the West. Additionally, prices are projected to peak in the fall and drop in winter, regardless of the scenario.
